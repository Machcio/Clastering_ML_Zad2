```python
from keras.datasets.mnist import load_data
(trainX,trainY), (testX,testY) = load_data()
```

```python
print(trainX.shape)
```

```python
from matplotlib import pyplot
```

```python
pyplot.imshow(trainX[1], cmap = 'gray_r')
```

```python
pyplot.subplot(221)

# equivalent but more general
ax1 = pyplot.subplot(2, 2, 1)

# add a subplot with no frame
ax2 = pyplot.subplot(222, frameon=False)

# add a polar subplot
pyplot.subplot(223, projection='polar')

# add a red subplot that shares the x-axis with ax1
pyplot.subplot(224, sharex=ax1, facecolor='red')

# delete ax2 from the figure
pyplot.delaxes(ax2)

# add ax2 to the figure again
pyplot.subplot(ax2)

# make the first axes "current" again
pyplot.subplot(221)
```

```python
def plotter(nrows,ncols):
  pyplot.figure(figsize=(20,20))
  for i in range(nrows*ncols):
    pyplot.subplot(nrows,ncols,i+1)
    pyplot.axis('off')
    pyplot.imshow(trainX[i], cmap = 'gray_r')
    pyplot.title(trainY[i])
```

```python
plotter(10,10)
```
